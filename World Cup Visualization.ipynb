```python
from matplotlib import pyplot as plt
import pandas as pd 
#import numpy as np
import seaborn as sns 

df = pd.read_csv("/Users/yumelhernandez/Downloads/fifa-world-cup/WorldCupMatches.csv")

```

```python
df["Total Goals"] = (df["Home Team Goals"] + df["Away Team Goals"])
print(df.head())
```

```python
sns.set_style("whitegrid")
sns.set_context("poster")
f, ax = plt.subplots(figsize=(12,7))
ax = sns.barplot(x="Year", y="Total Goals", data=df)
ax.set_title("FIFA total goals per year")
ax.set_xticklabels(df["Year"], rotation=90)
plt.show()


f, ax2 = plt.subplots(figsize=(8, 5))
ax2 = sns.boxplot(data=df, x="Year", y="Home Team Goals")
ax2.set_title("Goals per FIFA cup")
ax2.set_xticklabels(df["Year"], rotation=90)
plt.show()
```
